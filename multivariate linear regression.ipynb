```python
# Import numpy and pandas package 

import pandas as pd
import numpy as np 

# Data visualization

from matplotlib import pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
data = pd.read_csv("Housing.csv")
```

```python
data
```

```python
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
road_encoded=le.fit_transform(data['mainroad'])
furnish_encoded=le.fit_transform(data['furnishingstatus'])
data['mainroad'] = road_encoded
data['furnishingstatus'] = furnish_encoded
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
Y = data.price# includes the fields other than prices
X = data.iloc[:,:5]
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state=355)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
y_predict = regression.predict(x_test)
```

```python
y_predict
```

```python
plt.scatter(y_test,y_predict)
plt.xlabel('y_test', fontsize=18) 
plt.ylabel('y_pred', fontsize=16)
plt.plot(y_test,y_predict)
```
